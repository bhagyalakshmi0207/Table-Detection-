Resizing the image using Pillow library

```python
from PIL import Image
```

```python
img=Image.open('/content/dog.jpg')
img_resize=img.resize((200,200))
img_plot=plt.imshow(img_resize)
plt.show()
```

```python
img_resize.save('/content/dog_resized.jpg')
```

```python
# displaying the image from numpy array

img_res = mpimg.imread('/content/dog_resized.jpg')
img_res_plot = plt.imshow(img_res)
plt.show()
```

Converting RGB images to Grayscale image using OpenCV

```python
# importing OpenCV library
import cv2
```

```python
img =cv2.imread('/content/dog.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_img=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_img)
```

```python
grayscale_img.shape
```
